```python
# Transformações Lineares

import matplotlib.pyplot as plt
```

```python
# definindo o vetor (2, 2)
vetor = [2, 2]
```

```python
# definindo as origens para o vetor
origens = [(0, -1), (0, 1), (0, 0), (0, 2)]
```

```python
# extraindo as coodernadas x e y das origens
origens_x = [origem[0] for origem in origens]
origens_y = [origem[1] for origem in origens]
```

```python
# plotando o vetor(2, 2) com as origens especificadas
plt.figure()
plt.quiver(origens_x, origens_y, [vetor[0]]*len(origens_x), [vetor[1]]*len(origens_y), angles='xy', scale_units='xy', scale=1, color='g')

plt.xlim(-1, 5)
plt.ylim(-2, 5)
plt.xlabel('x')
plt.ylabel('y')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(color= 'gray', linestyle= '--', linewidth=0.5)
plt.gca().set_aspect('equal', adjustable='box')
plt.title("Um vetor com diferente pontos de origem")
plt.show()
```

```python
# Base canônica

# definindo o vetor de base canôninca

vetor_x = [1, 0] # vetor unitario ao longo do eixo x
vetor_y = [0, 1] # vetor unitario ao longo do eixo y 
```

```python
# plotando o vetor de base canônica
plt.figure()
plt.quiver(0, 0, vetor_x[0], vetor_x[1], angles='xy', scale_units='xy', scale=1, color='r', label='eixo x')
plt.quiver(0, 0, vetor_y[0], vetor_y[1], angles='xy', scale_units='xy', scale=1, color='b', label='eixo y')

plt.xlim(-1, 1)
plt.ylim(-1, 1)
plt.xlabel('x')
plt.ylabel('y')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(color= 'gray', linestyle= '--', linewidth=0.5)
plt.gca().set_aspect('equal', adjustable='box')
plt.legend()
plt.title("Base Canônica")
plt.show()
```

```python
# exemplo 2

vetor_x = [1, 0] 
vetor_y = [0, 1]

plt.figure()
plt.quiver(0, 0, vetor_x[0], vetor_x[1], angles='xy', scale_units='xy', scale=1, color='r', label='eixo x')
plt.quiver(0, 0, vetor_y[0], vetor_y[1], angles='xy', scale_units='xy', scale=1, color='b', label='eixo y')
plt.quiver(0, 0, 4, 3, angles='xy', scale_units='xy', scale=1, color='g', label='(4, 3)')

plt.xlim(-1, 5)
plt.ylim(-1, 5)
plt.xlabel('x')
plt.ylabel('y')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(color= 'gray', linestyle= '--', linewidth=0.5)
plt.gca().set_aspect('equal', adjustable='box')
plt.legend()
plt.title("Combinações feitas com os vetores da base canônica")
plt.show()
```
