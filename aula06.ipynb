```python
# Autovalor e Autovetor
import numpy as np
```

```python
# definindo a matriz A
A = np.array([[6, -1],
              [2, 3]])

# calculando os autovalores e autovetores
autovalores, autovetores = np.linalg.eig(A)

# criando a matriz diagonal com os autovalores
matriz_diagonal = np.diag(autovalores)

# imprimindo os autovalores
print("autovalores:")
print(autovalores)

# imprimindo os autovetores
print("\nautovetores:")
print(autovetores)

print("\nmatriz diagonal:\n", matriz_diagonal)
```

```python
# o primeiro autovalor é autovalores[0] e o autovetor associado é a coluna autovetores[:,0]
primeiro_autovalor = autovalores[0]
autovetor_associado = autovetores[:,0]

print("primeiro autovalor:", primeiro_autovalor)
print("autovetor associado ao primeiro autovalor:", autovetor_associado)

print(f"dividindo por {autovetor_associado[0]}, {autovetor_associado/autovetor_associado[0]}")

# o segundo autovalor é autovalores[1] e o autovetor associado é a coluna autovetores[:,1]
segundo_autovalor = autovalores[1]
autovetor_associado = autovetores[:,1]

print("\nsegundo autovalor:", segundo_autovalor)
print("autovetor associado ao segundo autovalor:", autovetor_associado)

print(f"dividindo por {autovetor_associado[0]}, {autovetor_associado/autovetor_associado[0]}")

```

```python
# reconstrução a matriz
# verificando a decomposição espectral

A_recontruida = autovetores @ matriz_diagonal @ np.linalg.inv(autovetores)
print("a reconstruida:\n", A_recontruida)
```

```python
# representação geométrica
import matplotlib.pyplot as plt

# definindo a matriz A
A = np.array([[6, -1], [2, 3]])

# calculando os autovalores e autovetores
autovetores, autovetores = np.linalg.eig(A)

# plotando os autovetores
origin = [0, 0] # origem

plt.quiver(*origin, *autovetores[:,0], color=['r'], scale=8, label=f'autovetor 1: {autovetores[:,0]}')
plt.quiver(*origin, *autovetores[:,1], color=['b'], scale=8, label=f'autovetor 2: {autovetores[:,1]}')

# configurando o grafico
plt.xlim(-3, 3)
plt.ylim(-3, 3)
plt.axvline(x=0, color='grey', lw=1)
plt.axhline(y=0, color='grey', lw=1)
plt.grid(True)
plt.legend()
plt.show()
```

```python
# autovetores calculado anteriormente
v1 = np.array([1, 1])
v2 = np.array([1, 2])

# plotando os autorevetores
plt.quiver(0, 0, v1[0], v1[1], angles='xy', scale_units='xy', scale=1, color='r', label='autovetor 1')
plt.quiver(0, 0, v2[0], v2[1], angles='xy', scale_units='xy', scale=1, color='b', label='autovetor 2')

# linhas que estendem os autovetores
t = np.linspace(-2, 2, 10)
for val in t:
    plt.plot(v1[0]*t, v1[1]*t, 'r--')
    plt.plot(v2[0]*t, v2[1]*t, 'b--')

# configurando o grafico
plt.xlim(-2, 2)
plt.ylim(-2, 2)
plt.axvline(x=0, color='grey', lw=1)
plt.axhline(y=0, color='grey', lw=1)
plt.grid(True)
plt.legend()
plt.show()
```
