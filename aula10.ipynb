```python
# produto misto e volume
import numpy as np
import matplotlib.pyplot as plt
```

```python
# exemplo 1

a = (1, 2, 6)
b = (5, 2, 1)
c = (2, 2, 10)

np.abs(np.dot(c, np.cross(a, b)))
```

```python
# exemplo 2

a = np.array([1, 2, 6])
b = np.array([5, 2, 1])
c = np.array([2, 2, 10])

np.abs(np.linalg.det(np.array([c, a, b])))
```

```python
# exercicio1

a = np.array([4, 9, -6])
b = np.array([5, 2, -2])
produto_escalar1 = np.dot(a, b)
print("produto escalar de a.b =", produto_escalar1)

# exercicio2

c = np.array([2, -3, 1])
d = np.array([0, 1, 4])
produto_escalar2 = np.dot(c, d)
print("produto escalar de c.d =", produto_escalar2)

# exercicio3

e = np.array([1, 2, 3])
f = np.array([-2, 1, 0])
produto_escalar3 = np.dot(e, f)
print("produto escalar de e.f =", produto_escalar3)
```

```python
# representação geometrica

# definindo os vetores
vetores = {
    "a": np.array([4, 9, -6]),
    "b": np.array([5, 0, -2]),
    "c": np.array([2, -3, 1]),
    "d": np.array([0, 1, 4]),
    "e": np.array([1, 2, 3]),
    "f": np.array([-2, 1, 0]),
    "g": np.array([3, 4, 5]),
    "h": np.array([1, -1, 2]),
    "i": np.array([-4, -7, -3]),
    "j": np.array([4, 9, -6])
}

# criando o grafico
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# plotando os vetores
for label, vetor in vetores.items():
    ax.quiver(0, 0, 0, vetor[0], vetor[1], vetor[2], label=label)

# configurações adicionais do grafico
ax.set_xlim([-10, 10])
ax.set_ylim([-10, 10])
ax.set_zlim([-10, 10])
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.legend()

# mostrando a figura
plt.show()
```

```python
# função para plotar vetores em 3D

def plot_vectors(v1, v2, label1, label2):
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')

    # plotando os vetores
    ax.quiver(0, 0, 0, v1[0], v1[1], v1[2],color='b', label=label1)
    ax.quiver(0, 0, 0, v2[0], v2[1], v2[2],color='r', label=label2)

    # configurações adicionair do grafico
    ax.set_xlim([-10, 10])
    ax.set_ylim([-10, 10])
    ax.set_zlim([-10, 10])
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')
    ax.legend()
    
    plt.show()

# exercicio1
a = np.array([4, 9, -6])
b = np.array([5, 2, -2])
plot_vectors(a, b, 'a', 'b')

# exercicio2
c = np.array([2, -3, 1])
d = np.array([0, 1, 4])
plot_vectors(c, d, 'c', 'd')

# exercicio3
e = np.array([1, 2, 3])
f = np.array([-2, 1, 0])
plot_vectors(e, f, 'e', 'f')

# exercicio4
g = np.array([3, 4, 5])
h = np.array([1, -1, 2])
plot_vectors(g, h, 'g', 'h')

# exercicio5
i = np.array([-4, -7, -3])
j = np.array([4, 9, -6])
plot_vectors(i, j, 'i', 'j')
```

```python
import plotly.graph_objs as go
```

```python
# funcção para plotar vetores em 3D com plotly
def plot_vectors_plotly(v1, v2, label1, label2):
    fig = go.Figure()

    # adicionando os vetores ao grafico
    fig.add_trace(go.Scatter3d(x=[0, v1[0]], y=[0, v1[1]], z=[0, v1[2]], mode='lines+markers', name=label1))
    fig.add_trace(go.Scatter3d(x=[0, v2[0]], y=[0, v2[1]], z=[0, v2[2]], mode='lines+markers', name=label2))

    # configurando o layout do grafico
    fig.update_layout(scene=dict(
        xaxis=dict(title='X'),
        yaxis=dict(title='Y'),
        zaxis=dict(title='Z')),
        margin=dict(l=0, r=0, b=0, t=0))
    
    fig.show()
```

```python
# exercicio1
a = np.array([4, 9, -6])
b = np.array([5, 0, -2])
plot_vectors_plotly(a, b, 'a', 'b')

# exercicio2
c = np.array([2, -3, 1])
d = np.array([0, 1, -4])
plot_vectors_plotly(c, d, 'c', 'd')

# exercicio3
e = np.array([1, 2, 3])
f = np.array([-2, 1, 0])
plot_vectors_plotly(e, f, 'e', 'f')
```
