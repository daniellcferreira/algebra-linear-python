```python
# Bases
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plot_vectors():
    e1 = np.array([1, 0])
    e2 = np.array([0, 1])

    plt.figure()
    plt.quiver(0, 0, e1[0], e1[1], angles='xy', scale_units='xy', scale=1, color='r')
    plt.quiver(0, 0, e2[0], e2[1], angles='xy', scale_units='xy', scale=1, color='b')

    plt.xlim(-0.25, 1.25)
    plt.ylim(-0.25, 1.25)
    plt.grid()
    plt.axhline(0, color='black', linewidth=0.5)
    plt.axvline(0, color='black', linewidth=0.5)
    plt.gca().set_aspect('equal', adjustable='box')

    plt.text(e1[0], e1[1], 'e1', fontsize=12, ha='right')
    plt.text(e2[0], e2[1], 'e2', fontsize=12, ha='right')

    plt.show()

plot_vectors()
```

```python

def plot_vectors(ax, v, u, label_v, label_u, op=None, label_vpu=None):
    # Plotando os vetores v e u
    ax.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, color='r')
    ax.quiver(0, 0, u[0], u[1], angles='xy', scale_units='xy', scale=1, color='b')
    
    # Adicionando labels para v e u
    ax.text(v[0], v[1], label_v, fontsize=12, ha='right')
    ax.text(u[0], u[1], label_u, fontsize=12, ha='right')
    
    if op == '+':
        # Calculando e plotando a soma dos vetores
        vpu = np.add(v, u)
        ax.quiver(0, 0, vpu[0], vpu[1], angles='xy', scale_units='xy', scale=1, color='g')
        ax.text(vpu[0], vpu[1], label_vpu, fontsize=12, ha='right')
    
    # Configurando o gráfico
    ax.set_xlim(-0.5, 2.5)
    ax.set_ylim(-0.5, 2.5)
    ax.grid()
    ax.axhline(0, color='black', linewidth=0.5)
    ax.axvline(0, color='black', linewidth=0.5)
    ax.set_aspect('equal', adjustable='box')

# Criando a figura com 3 subplots
fig, axs = plt.subplots(1, 3, figsize=(8, 4))

# Primeiro subplot
plot_vectors(axs[0], v=np.array([1, 0]), u=np.array([0, 2]), label_v='$e_1$', label_u='$2e_2$')

# Segundo subplot
plot_vectors(axs[1], v=np.array([1, 0]), u=np.array([0, 2]), label_v='$e_1$', label_u='$2e_2$', op='+', label_vpu='$e_1 + 2e_2$')

# Terceiro subplot
plot_vectors(axs[2], v=np.array([1, 2]), u=np.array([0, 0]), label_v='$e_1 + 2e_2$', label_u='')

# Ajustando layout e mostrando o gráfico
plt.tight_layout()
plt.show()
```
